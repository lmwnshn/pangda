```python
from sdv.tabular import CTGAN
from sdv.constraints import Unique

import warnings
import pandas as pd

data = pd.DataFrame({'id': range(10), 'id2': range(10)})

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    model = CTGAN(constraints=[Unique(columns=['id', 'id2'])])
    model.fit(data)
    sample = model.sample(num_rows=1000)
    print(sample.shape)
    display(sample)
```

```python
from sdv.tabular import CTGAN
from sdv.constraints import Unique

import warnings
import pandas as pd

data = pd.DataFrame({'id': ['asfasa','fasdfasb','asdasd','asdasdsa','fasfsa','safasds']})

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    model = CTGAN()
    model.fit(data)
    sample = model.sample(num_rows=10)
    print(sample.shape)
    display(sample)
```
